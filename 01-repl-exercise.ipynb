```python
from utils import claude
from anthropic.types import MessageParam

messages: list[MessageParam] = []

while True:
    user_input = input("> ")
    print(">", user_input)

    if user_input == "exit":
        break

    claude.add_user_message(messages, user_input)
    answer = claude.chat(messages)
    claude.add_assistant_message(messages, answer)

    print(answer)
```
